```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/drive/My Drive/FatimaQurban_ML&DL_BytewiseFellowship2024/FatimaQurban_ML&DL_Task2/netflix_data.csv")
print(df.shape)
df.head()
```

```python
# Handling null values
df = df.dropna()  # Drops all rows with any null values
print(df.shape)
```

```python
# bool array to filter the type movie
is_movie=df['type']=="Movie"

# store
netflix_subset=df[is_movie]

other_column=['title','country','genre','date_added','release_year', 'duration']

netflix_movies=netflix_subset[other_column]

netflix_movies.head()
```

```python
short_movies=netflix_movies[netflix_movies['duration']<=60]
short_movies.head()
```

```python
colors=[]

for index, row in netflix_movies.iterrows():
  if 'Children' in row['genre']:
      colors.append('blue')
  elif 'Documentaries' in row['genre']:
      colors.append('green')
  elif 'Stand-Up' in row['genre']:
      colors.append('red')
  else:
      colors.append('black')

```

```python
fig = plt.figure(figsize=(12,8))
plt.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=colors)
plt.title('Movie Duration by Year of Release')
plt.xlabel('Release Year')
plt.ylabel('Duration (min)')

plt.show()
```
