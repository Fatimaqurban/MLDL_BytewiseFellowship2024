```python
import pandas as pd
import matplotlib as plt
import numpy as np
```

```python
df=pd.read_csv("/content/drive/My Drive/FatimaQurban_ML&DL_BytewiseFellowship2024/FatimaQurban_ML&DL_Task2/schools.csv")
print(df.shape)
df.head()
```

# PART A

```python
df['average_math_percent']=(df['average_math']/800)*100
sorted_df=df.sort_values(by='average_math_percent', ascending=False)
best_math_percentages= sorted_df['average_math_percent'] >= 80

best_math_school = sorted_df.loc[best_math_percentages, 'average_math_percent']
school_names= sorted_df.loc[best_math_percentages, 'school_name']

print("Best Math Schools and their Average Math Percentages:\n")
for name, percent in zip(school_names, best_math_school):
    print(name, percent)
```


# PART B

```python

df['total_SAT']= df['average_math'] + df['average_reading'] + df['average_writing']
new_df=df[['school_name', 'total_SAT']]
top_ten_schools_sort= new_df.sort_values(by='total_SAT', ascending=False)
top_ten_schools=top_ten_schools_sort.head(10)
print(top_ten_schools)
```

# PART C

```python
grouping_borough= df.groupby('borough')['total_SAT']
for borough, total_SAT_scores in grouping_borough:
    print(f"{borough} Group: {list(total_SAT_scores)}")

```

```python
average_SAT= grouping_borough.mean()
std_SAT=grouping_borough.std()
num_schools=grouping_borough.count()

max_std_borough = std_SAT.idxmax()

largest_std_dev_info={
    "Borough":max_std_borough,
    "No of schools":num_schools[max_std_borough],
    "SAT score":round(average_SAT[max_std_borough],2),
    "Standrard deviation":round(std_SAT[max_std_borough],2)
}
largest_std_dev=pd.DataFrame([largest_std_dev_info])
print("Information of the Borough with the largestest standard deviation \n")
print(largest_std_dev)
```
