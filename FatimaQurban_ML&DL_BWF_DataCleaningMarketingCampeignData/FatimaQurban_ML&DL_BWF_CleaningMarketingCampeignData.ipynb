# Cleaning Client Dataframe

```python
# df_client['job'] = df_client['job'].replace("unknown", np.NaN)
# df_client['education'] = df_client['education'].replace("unknown", np.NaN)

#the above two rows are not changing the original dataframe but making a copy of the row, so to make a change in the original dataframe we use
df_client.loc[:, "job"]= df_client['job'].replace("unknown", np.NaN)
df_client.loc[:, "education"]= df_client['education'].replace("unknown", np.NaN)

```

```python
df_client.sample(10)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

ss= StandardScaler()
le= LabelEncoder()

df_client.loc[:,"credit_default"] = le.fit_transform(df_client["credit_default"])
df_client.loc[:,"mortgage"] = le.fit_transform(df_client["mortgage"])
df_client.loc[:,"job"] = le.fit_transform(df_client["job"])
df_client.loc[:,"marital"] = le.fit_transform(df_client["marital"])
df_client.loc[:,"education"] = le.fit_transform(df_client["education"])

df_client.loc[:,"age"] = ss.fit_transform(df_client["age"].values.reshape(-1, 1))
df_client.loc[:, "age"] = df_client["age"].values.flatten()

```

```python
df_client.sample(10)
```

# Cleaning Economic Dataframe

---



```python
from sklearn.preprocessing import StandardScaler

ss= StandardScaler()
# Reshape the columns to 2D and apply StandardScaler as standard scaler acccept 2d array
df_economic.loc[:, "cons_price_idx"] = ss.fit_transform(df_economic["cons_price_idx"].values.reshape(-1, 1))
df_economic.loc[:, "euribor_three_months"] = ss.fit_transform(df_economic["euribor_three_months"].values.reshape(-1, 1))

# If you want to reshape it back to 1D, you can use .flatten()
df_economic.loc[:, "euribor_three_months"] = df_economic["euribor_three_months"].values.flatten()
df_economic.loc[:, "cons_price_idx"] = df_economic["cons_price_idx"].values.flatten()

```

```python
df_economic.sample(10)
```

# Cleaning Campaign Dataframe

---



```python
df_campaign.loc[:, "previous_campaign_contacts"] = le.fit_transform(df_campaign["previous_campaign_contacts"])
df_campaign.loc[:, "previous_outcome"] = le.fit_transform(df_campaign["previous_outcome"])
df_campaign.loc[:, "campaign_outcome"] = le.fit_transform(df_campaign["campaign_outcome"])
```

```python
month_conversion = {'jan': '01', 'feb': '02', 'mar': '03', 'apr': '04', 'may': '05', 'jun': '06',
                    'jul': '07', 'aug': '08', 'sep': '09', 'oct': '10', 'nov': '11', 'dec': '12'}

# Apply the conversion
df_campaign.loc[:,'month'] = df_campaign['month'].replace(month_conversion)


# Convert 'day' and 'month' columns to strings
df_campaign.loc[:,'day'] = df_campaign['day'].astype(str).str.zfill(2)
df_campaign.loc[:,'month'] = df_campaign['month'].astype(str).str.zfill(2)

# Combine 'day' and 'month' into a datetime column with a placeholder year
df_campaign.loc[:,'datetime'] = pd.to_datetime(df_campaign['month'] + '-' + df_campaign['day'], format='%m-%d')

df_campaign.insert(5, 'datetime', df_campaign.pop('datetime'))
df_campaign=df_campaign.drop(columns=["month"])
df_campaign=df_campaign.drop(columns=["day"])
```
