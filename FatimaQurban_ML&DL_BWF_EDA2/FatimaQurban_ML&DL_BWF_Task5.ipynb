```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/drive/My Drive/FatimaQurban_ML&DL_BytewiseFellowship2024/FatimaQurban_ML&DL_BWF_Task5/DATASET.csv')
df.head()
```

### PART #1

```python
df['TIME OCC']= df['TIME OCC'].astype(str) #  converting to string to to slice the column later
df['HOUR OCC']= df['TIME OCC'].str[:2].astype(int)
peak_crime_hour=df['HOUR OCC'].value_counts().idxmax()
print("Peak Crime Hour: ", peak_crime_hour)
```

### PART #2

```python
# peak_night_crime_location

df['TIME OCC']= df['TIME OCC'].astype(str)
df['HOUR OCC']=df['TIME OCC'].str[:2].astype(int)
# til_midnight= df['HOUR OCC']>22 && df['HOUR OCC']<24
crime_hours=(df['HOUR OCC']>22) | (df['HOUR OCC']<5)
peak_hours=df[crime_hours]
peak_night_crimes = peak_hours['AREA NAME'].value_counts()
print("peak night crimes and their location ", peak_night_crimes)
```

```python
peak_night_crime_location= peak_night_crimes.idxmax()
print("peak night crime location is ", peak_night_crime_location)
```

### PART # 3

```python
# Define bins and labels
bins = [0, 17, 25, 34, 44, 54, 64, float('inf')]
labels = ["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]

df['Age Category'] = pd.cut(df['Vict Age'], bins=bins, labels=labels)
print(df['Age Category'])

```

```python
age_crime_freq = df.groupby('Age Category').size()

# Create a pandas Series with age group labels as index and crime frequencies as values
victim_ages = pd.Series(age_crime_freq, index=labels)
print(victim_ages)
```
