-----------------------------------------------------------------------------------------------------**Question # 4**-------------------------------------------------------------------------------------------

```python
import numpy as np

# Lists of dictionaries
players =[
      {"player_name":"A", "height_in":74, "weight_lb":150},
      {"player_name":"B", "height_in":77, "weight_lb":154},
      {"player_name":"C", "height_in":75, "weight_lb":160},
      {"player_name":"D", "height_in":76, "weight_lb":200},
      {"player_name":"E", "height_in":72, "weight_lb":160},

]

bmi_list=[]

print("\n")

for player in players:
  np_height_in =  np.array(player["height_in"])*0.0254
  np_weight_lb= np.array(player["weight_lb"])*0.4535
  bmi= np_weight_lb/np.square(np_height_in)
  print("Player Name:",player['player_name'], ", BMI:", bmi)

  #since its a dict so bmi is not an np array but an var which has last element so we have to create a new list to append then convert to numpy
  bmi_list.append(bmi)

print('\n')

#convert the list to numpy array
bool_bmi_array=np.array(bmi_list)
print('Players BMI:', bool_bmi_array)

# iterate the numpy array to val less than 21 and store in an bool array 'Light'
Light=bool_bmi_array<21    #--------> only np array can allow this, Cant do this to normal list

print('Players BMI < 21 boolen Array(Light):', Light)
print('Players BMI < 21;',bool_bmi_array[Light])
```

```python
import numpy as np

positions = ['GK', 'M', 'A', 'D', 'GK', 'M', 'A', 'D']
heights = [191, 184, 185, 180, 192, 175, 178, 182]

np_position= np.array(positions)
np_height= np.array(heights)

GK_height=(np_height[np_position=='GK'])
otherPlayer_height=(np_height[np_position!='GK'])

print('Median Height of the GK:',np.median(GK_height))
print('Median Height of other players:', np.median(otherPlayer_height))

print('Hence proved, GK median height is higher than other players!')
```
