## Function with args
You can add multiple arguments using args in the parameter

```python
def StudentCourses(name, *args):
  courses=', '.join(args) #this .join() removes the bracket which we get from the use of args
  print(f'{name} takes courses {courses}')

StudentCourses("Alina", "maths","physics","computer")
```

## Function with kwargs(Keyword Argument)
with kwargs we are passing argument as key value pair, we can pass multiple key value pairs

### Thus is the **list comprehension** method:
[expression **for** item **in** iterable **if** condition]



```python
def personal_details(name, *args,**kwargs):
  friends= ", ".join(args)
  courses=" and ".join([f"{course} course with {details}" for course, details in kwargs.items()])

  print(f" {name} has friends {friends}. She takes {courses}")

personal_details("Ayesha", "Ali","Natasha", "Mona", English="1 credit hour", Programming="3 credit hour")
```

**Alternative way** to iterate the key value pair (Kwargs) is:

```python
def student_details(name, *args, **kwargs):
  friend=", ".join(args)

  courses_list= []
  for course, details in kwargs.list():
    courses_list.append(f"{course} course with {details}")
    course=" and ".join(courses_list)
    print(f"{name} has {friend} amd she takes {course}")

personal_details("Ayesha", "Ali","Natasha", "Mona", English="1 credit hour", Programming="3 credit hour")
```

# Lambda
Anonymus function that can have any number of arguments but can have only one expression.

**Syntax:**

*   [**lambda** arguments: expression]




```python
def multiply(x,y):
  return x*y
print(f"Simple multiplication: {multiply(10,20)}")

multiple_lambda= lambda x,y: x*y
print(f"Lambda multiplication: {multiple_lambda(20,20)}")
```

## Uses Of Lambda In Machine Learning:

### Feature Engineering

```python
import pandas as pd

df= pd.DataFrame({
    'Age': [25, 30, 35],
    'Salary': [50000, 60000, 70000]
})

df["Position"]= df['Salary'].apply(lambda x: 'Senior' if x>60000 else 'Junior')
print(df)
```

### Data Cleaning

```python
import numpy as np

df= pd.DataFrame({
    'A':[1,2,np.nan,4],
    'B':[4,5,6,np.nan]
})

cleaned_df= df.apply(lambda x: x.fillna(x.mean().astype(int)))
print(cleaned_df)
```
